```python
import numpy as np

# Sizes of 4 different plots in square feet
plot_sizes = np.array([2250, 4500, 1125, 9000]) 

# Base price per square foot in PKR
base_prices = np.array([6500, 8000, 5500, 9500])

# Sector premium multipling factor (e.g., premier sectors add a cost multiplier)
sector_premiums = np.array([1.1, 1.25, 1.0, 1.35])

# calculate the base price of the plot
raw_value = plot_sizes * base_prices

# calculate the final price of the plot
final_value = raw_value * sector_premiums

# format all floats with commas and 2 decimal places globally
np.set_printoptions(formatter={"float": lambda x: f"{x:,.2f} PKR"})

print(final_value)
print(f"Total Investement Portfolio Value is {final_value.sum():,.2f} PKR")
```
