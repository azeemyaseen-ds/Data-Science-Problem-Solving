```python
# the raw data containing list of raw string representing each server log line
server_logs = [
    "[2026-06-22 10:00:01] INFO - AuthModule - User logged in successfully.",
    "[2026-06-22 10:05:23] ERROR - PaymentModule - Gateway timeout occurred.",
    "[2026-06-22 10:06:12] WARNING - DatabaseModule - High memory usage detected.",
    "[2026-06-22 10:10:45] ERROR - AuthModule - Invalid token signature.",
    "[2026-06-22 10:12:00] ERROR - PaymentModule - Card verification failed."
]

def miner_server_log(logs):
    total_errors = 0
    error_by_system = {}

    for log in logs:
        # 1. split the string intentionally using the " - " delimeter
        # it can split the string into 3 clean parts
        parts = log.split(" - ")

        # 2. extract the log level from the end of the first part
        first_part = parts[0]
        
        if "ERROR" in first_part:
            total_errors += 1
            module_name = parts[1].strip()
        
            # calculate ERROR message by each system module
            error_by_system[module_name] = error_by_system.setdefault(module_name, 0) + 1
    
    print(f"There are {total_errors} occured in the whole system!")
    print(f"Error occured per module: {error_by_system}")

miner_server_log(server_logs)
```
