```python
# import the necessary libraries use in the project
import pandas as pd
import numpy as np

# raw data
raw_ledger = {
    "Transaction_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "City": ["Islamabad", "Karachi", "Islamabad", "Rawalpindi", "Karachi", "Lahore", "Islamabad", "Rawalpindi", "Lahore", "Karachi"],
    "Product_Category": ["Electronics", "Clothing", "Electronics", "Home", "Electronics", "Clothing", "Home", "Electronics", "Home", "Clothing"],
    "Units_Sold": [2, 10, 1, 4, 5, 3, 8, 2, 1, 12],
    "Unit_Price_PKR": [45000, 1200, 35000, 8000, 55000, 2500, 12000, 60000, 4500, 1500]
}

# convert the raw data into a pandas dataframe to perform vectorized operations
df_ledger = pd.DataFrame(raw_ledger)

# display the first five rows of data
df_ledger.head()
```

```python
# get the overview of the data
df_ledger.info()
```

#### 1. **Total Transaction Revenue:** Create a new column named `Total_Revenue` calculated by multiplying units sold by unit price.

```python
df_ledger['Total_Revenue'] = df_ledger['Units_Sold'] * df_ledger['Unit_Price_PKR']

# verify the result
df_ledger.head(3)
```

#### 2. **City-Wise Analytics:** Generate a summary showing the **Total Revenue** and **Average Units Sold** per individual city.

```python
# use the groupby() method with agg() method to generate the summary
city_summary = df_ledger.groupby("City").agg(
    total_revenue = ("Total_Revenue", "sum"),
    avg_units_sold = ("Units_Sold", "mean")
)

city_summary['avg_units_sold'] = city_summary['avg_units_sold'].round(2)

city_summary
```

#### 3. **The Revenue Outlier Flags:** Identify individual transactions where the `Total_Revenue` is strictly **above the overall platform average transaction revenue**. Filter and display a dataframe showing only these premium, high-value transactions.

```python
avg_trans_revenue = df_ledger['Total_Revenue'].mean()

premium_trans_df = df_ledger[df_ledger['Total_Revenue'] > avg_trans_revenue]

premium_trans_df
```

#### 4. **Category Market Share:** Calculate what percentage of total platform revenue belongs to each unique `Product_Category`.

```python
category_revenue = df_ledger.groupby("Product_Category")["Total_Revenue"].sum()

market_share = (category_revenue / category_revenue.sum()) * 100

market_share_df = market_share.round(2).to_frame(name="market_share_percentage")

market_share_df
```
