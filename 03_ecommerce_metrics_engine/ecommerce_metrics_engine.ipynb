```python
sales_orders = [
    {"order_id": 1001, "category": "Electronics", "price": 12000, "quantity": 1},
    {"order_id": 1002, "category": "Clothing", "price": 2500, "quantity": 3},
    {"order_id": 1003, "category": "Electronics", "price": 15000, "quantity": 1},
    {"order_id": 1004, "category": "Home Decor", "price": 4000, "quantity": 2},
    {"order_id": 1005, "category": "Clothing", "price": 1500, "quantity": 2}
]

def generate_sales_metrics(orders):
    total_revenue = 0
    revenue_by_category = {}

    # We can use len function to track total number of orders
    total_orders = len(orders)

    for order in orders:

        # 1. Calculate the total revenue by specific transaction
        revenue_per_order = order['price'] * order['quantity']
        total_revenue += revenue_per_order
        
        category = order['category']
        # 2. Calculate total revenue by each category using a dictionary
        if category not in revenue_by_category:
            revenue_by_category[category] = revenue_per_order
        else:
            revenue_by_category[category] += revenue_per_order
    
    # figure out the best performing product category
    best_category = None
    highest_revenue = 0

    for category, revenue in revenue_by_category.items():
        if revenue > highest_revenue:
            highest_revenue = revenue
            best_category = category
    
    # 3. Calculate the average order value (AOV)
    average_order_value = total_revenue / total_orders if total_orders > 0 else 0
    
    # results
    print(f"The Total Revenue of Entire Platform is {total_revenue} PKR")
    print(f"The Average Order Value is {average_order_value:.2f} PKR")
    print(f"Revenue by Category: {revenue_by_category}")
    print(f"Best Performing Product Category is {best_category} & their Revenue is {highest_revenue} PKR")

generate_sales_metrics(sales_orders)
```
