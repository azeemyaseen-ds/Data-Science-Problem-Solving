```python
# Raw Data remains the same
package_batch = [
    "ID-9082", 2.5, "Status: Pending", 
    "ID-1104", 12.0, "Status: Delivered", 
    "ID-3342", "6.8", "Status: Pending",
    "ID-5561", -1.2, "Status: Pending" 
]

def calculate_batch_shipping(batch, rate_per_kg=300):
    """
    Processes a logistics batch list, fixes corrupted weights,
    and returns the total shipping cost for pending packages.
    """
    total_cost = 0

    # step through the list 3 elements at a time
    for i in range(0, len(batch), 3):
        pkg_id = batch[i]
        weight = batch[i + 1]
        status = batch[i + 2]
    
        # 1. skip if already "Delieverd"
        if status != "Status: Pending":
            continue

        # 2. Fix Data Corruption: String conversion
        if isinstance(weight, str):
            weight = float(weight)
        
        # 3. Fix Data Corruption: Negative weights
        if weight < 0:
            print(f"TROUBLE SHOOTING: Negative weight detected for Package {pkg_id}")
            weight = abs(weight)
        
        # 4. Perform calculation once per valid package
        pkg_cost = weight * rate_per_kg
        print(f"The shipping cost for package {pkg_id} is {pkg_cost:.2f} PKR")

        total_cost += pkg_cost
    
    print(f"Grand Total Shipping Cost: {total_cost:.2f} PKR")
    return total_cost

# run the function
calculate_batch_shipping(package_batch)
```
