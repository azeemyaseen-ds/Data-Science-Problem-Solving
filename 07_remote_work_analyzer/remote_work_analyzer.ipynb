```python
# import the pandas library to perform vectorized operations on column values
import pandas as pd

survey_data = {
    "Employee": ["Ahmad", "Ayesha", "Azeem", "Sana", "Bilal", "Zainab"],
    "City": ["Islamabad", "Lahore", "Islamabad", "Rawalpindi", "Islamabad", "Karachi"],
    "Work_Type": ["Office", "Remote", "Remote", "Remote", "Office", "Remote"],
    "Monthly_Salary_PKR": [180000, 250000, 320000, 150000, 90000, 280000]
}

# convert the survey dictionary into a pandas dataframe
df = pd.DataFrame(survey_data)
df.head()
```

```python
# over view of the data
df.info()
```

```python
# filter the dataframe to isolate rows where the city == "Islamabad" & work_type == "Remote"
filtered_df = df[(df['City'] == "Islamabad") & (df['Work_Type'] == "Remote")]
```

```python
# average salary of those specific remote worker in islamabad
filtered_df['Monthly_Salary_PKR'].mean()
```

```python
# find the index of the row where "Monthly_Salaary_PKR" is highest within the filtered_df
highest_salary_index = filtered_df['Monthly_Salary_PKR'].idxmax()

# extract the "Employee" value at that specific row index
highest_paid_employee = filtered_df.loc[highest_salary_index, "Employee"]

print(f"The highest paid remote workers in Islamabad is: {highest_paid_employee}")
```
