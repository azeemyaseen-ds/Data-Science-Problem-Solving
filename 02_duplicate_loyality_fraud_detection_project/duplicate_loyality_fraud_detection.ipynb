```python
customer_records = [
    {"account_id": "ACC-101", "name": "  Azeem khan ", "phone": "+92-300-1234567", "points": 150},
    {"account_id": "ACC-102", "name": "AZEEM KHAN", "phone": "03001234567", "points": 50},
    {"account_id": "ACC-103", "name": "Zainab Bibi", "phone": "0321-9876543", "points": 400},
    {"account_id": "ACC-104", "name": "azeem   khan", "phone": "+923001234567", "points": 20}
]

def detect_loyality_fraud(records):
    # create tracking sets
    seen_names = set()
    seen_phones = set()

    # track bad credentials we find
    fraud_names = set()
    fraud_phones = set()

    # first phase: Clean the data and figure out the duplicate names and phones
    for record in records:

        # clean names (lowercase, strip, remove the double spaces between words)
        clean_name = " ".join(record['name'].lower().split())
        # clean phone (remove dashes, remove outer spaces, extract last 10 digits)
        clean_phone = record['phone'].replace("-", "").strip()[-10:]
        
        # save the clean version back to the record to use them later
        record['name'] = clean_name
        record['phone'] = clean_phone

        # check if we seen this name or phone before
        if record['name'] in seen_names:
            fraud_names.add(record['name'])
        if record['phone'] in seen_phones:
            fraud_phones.add(record['phone'])
        
        # add to our checklist
        seen_names.add(record['name'])
        seen_names.add(record['phone'])

    # second phase: calculate the total frozen point using our fraud list
    total_frozen_point = 0

    for record in records:
        # if this record uses a name or phone that is flagged as duplicate
        if record['name'] in fraud_names or record['phone'] in fraud_phones:
            print(f"⚠️ ACCOUNT LOCKED: {record['account_id']} | {record['name']} | {record['phone']} | {record['points']}")
            total_frozen_point += record['points']
        
    print(f"\n🔒 Total Freeze Points under Investigation: {total_frozen_point}")
    return total_frozen_point

detect_loyality_fraud(customer_records)
```
