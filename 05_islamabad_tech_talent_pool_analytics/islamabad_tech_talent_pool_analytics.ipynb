```python
applicant_profiles = [
    "Name: Ayesha | City: Islamabad | Skill: Python | Score: 85",
    "Name: Ahmad | City: Lahore | Skill: Java | Score: 90",
    "Name: Azeem | City: Islamabad | Skill: Python | Score: 95",
    "Name: Sana | City: Rawalpindi | Skill: Python | Score: 78",
    "Name: Bilal | City: Islamabad | Skill: C++ | Score: 60"
]

def talent_analytics_pool(profiles):
    matching_names = []
    total_matching_applicants = 0
    total_score_match_applicants = 0
    applicants_by_city = {}

    # Step 1: split string using a delimeter
    # use dictionary comprehension to create key value pairs using split parts
    # clean the trailling spaces around the words
    for profile in profiles:

        parts = profile.strip().split(" | ")

        data = {item.split(":")[0].strip(): item.split(":")[1].strip() for item in parts}

        city = data['City']
        skill = data['Skill']
        
        # Step 2: Identify applicants who live in Islamabad and posses the skill Python
        if city == "Islamabad" and skill == 'Python':

            # store the names of matching candidates
            matching_names.append(data['Name'])

            # count the no of matching candidates
            total_matching_applicants += 1
            
            # calculate their total total test score
            total_score_match_applicants += int(data['Score'])
        
        # Step 3: Track how many applicants are from different cities
        applicants_by_city[city] = applicants_by_city.setdefault(city, 0) + 1
    
    average_score = total_score_match_applicants / total_matching_applicants if total_matching_applicants > 0 else 0

    print(f"Applicants names who match the criteria (Islamabad + Python): {matching_names}")
    print(f"Average score of matching applicants {average_score}")
    print(f"Applicants by City: {applicants_by_city}")

talent_analytics_pool(applicant_profiles)
```
